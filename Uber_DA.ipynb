```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('/content/UberDataset (1).csv')
```

```python

df['START'].unique()
```

```python
df.drop(df[df['START_DATE'] == 'Totals'].index, inplace=True)
df['START_DATE'] = pd.to_datetime(df['START_DATE'], format='%m-%d-%Y %H:%M', errors='coerce')
df['END_DATE'] = pd.to_datetime(df['END_DATE'], format='%m-%d-%Y %H:%M', errors='coerce')

df['Duration'] = (df['END_DATE'] - df['START_DATE']).dt.total_seconds() / 60
df.rename(columns={'Duration': 'Duration_Minutes'}, inplace=True)

print(df[['START_DATE', 'END_DATE', 'Duration_Minutes']].head())
df.info()
```

```python
print('Descriptive Statistics for Miles:')
print(df['MILES'].describe())
print('\nDescriptive Statistics for Duration_Minutes:')
print(df['Duration_Minutes'].describe())

df.info()
```

```python
# Average miles and duration per purpose
purpose_analysis = df.groupby('PURPOSE')[['MILES', 'Duration_Minutes']].mean().sort_values(by='MILES', ascending=False)
print('\nAverage Miles and Duration per Purpose:')
print(purpose_analysis)
```

```python
trip_purpose=df['PURPOSE'].value_counts()
```

```python
# Visualization: Bar plot for trip purposes
plt.figure(figsize=(12, 6))
sns.barplot(x=trip_purpose.index, y=trip_purpose.values,)
plt.title('Top 10 Most Frequent Trip Purposes')
plt.xlabel('Purpose')
plt.ylabel('Number of Trips')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#trip category
category= df['CATEGORY'].value_counts()
print('Trip Category Distribution:')
print(category)
```

```python
# Average miles and duration per category
category_analysis = df.groupby('CATEGORY')[['MILES', 'Duration_Minutes']].mean()
print('\nAverage Miles and Duration per Category:')
print(category_analysis)
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(x='CATEGORY', data=df, palette='viridis')
plt.title('Distribution of Trip Categories')
plt.xlabel('Category')
plt.ylabel('Number of Trips')
plt.show()
```

```python
# Top 10 Start Locations
start_location_counts = df['START'].value_counts().head(10)
print('Top 10 Start Locations:')
print(start_location_counts)
```

```python
# Top 10 Stop Locations
stop_location_counts = df['STOP'].value_counts().head(10)
print('\nTop 10 Stop Locations:')
print(stop_location_counts)
```

```python
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.barplot(x=start_location_counts.index, y=start_location_counts.values, palette='Blues_d')
plt.title('Top 10 Most Frequent Start Locations')
plt.xlabel('Start Location')
plt.ylabel('Number of Trips')
plt.xticks(rotation=45, ha='right')
```

```python
plt.subplot(1, 2, 2)
sns.barplot(x=stop_location_counts.index, y=stop_location_counts.values, palette='Reds_d')
plt.title('Top 10 Most Frequent Stop Locations')
plt.xlabel('Stop Location')
plt.ylabel('Number of Trips')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['MILES'], kde=True, bins=50)
plt.title('Distribution of Trip Miles')
plt.xlabel('Miles')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(df['Duration_Minutes'], kde=True, bins=50)
plt.title('Distribution of Trip Duration (Minutes)')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Box plots for MILES and Duration_Minutes by CATEGORY
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='CATEGORY', y='MILES', data=df)
plt.title('Miles by Trip Category')
plt.xlabel('Category')
plt.ylabel('Miles')

plt.subplot(1, 2, 2)
sns.boxplot(x='CATEGORY', y='Duration_Minutes', data=df)
plt.title('Duration (Minutes) by Trip Category')
plt.xlabel('Category')
plt.ylabel('Duration (Minutes)')
plt.tight_layout()
plt.show()
```

```python
# Box plots for MILES and Duration_Minutes to identify outliers
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.boxplot(y=df['MILES'])
plt.title('Box Plot of Trip Miles')
plt.ylabel('Miles')

plt.subplot(1, 2, 2)
sns.boxplot(y=df['Duration_Minutes'])
plt.title('Box Plot of Trip Duration (Minutes)')
plt.ylabel('Duration (Minutes)')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot for MILES vs Duration_Minutes
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Duration_Minutes', y='MILES', data=df)
plt.title('Relationship between Trip Miles and Duration')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Miles')
plt.show()
```

```python
# Violin plots for MILES and Duration_Minutes by TOP 5 PURPOSES
top_purposes = df['PURPOSE'].value_counts().head(5).index
df_top_purposes = df[df['PURPOSE'].isin(top_purposes)]

plt.figure(figsize=(16, 7))
plt.subplot(1, 2, 1)
sns.violinplot(x='PURPOSE', y='MILES', data=df_top_purposes, palette='tab10')
plt.title('Miles Distribution by Top 5 Trip Purposes')
plt.xlabel('Purpose')
plt.ylabel('Miles')
plt.xticks(rotation=45, ha='right')

plt.subplot(1, 2, 2)
sns.violinplot(x='PURPOSE', y='Duration_Minutes', data=df_top_purposes, palette='tab10')
plt.title('Duration (Minutes) Distribution by Top 5 Trip Purposes')
plt.xlabel('Purpose')
plt.ylabel('Duration (Minutes)')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# monthly trndes
import calendar

df['MONTH'] = df['START_DATE'].dt.month_name()

plt.figure(figsize=(12,6))
month_order = list(calendar.month_name)[1:]
sns.countplot(data=df, x='MONTH', order=month_order, palette='plasma')
plt.xticks(rotation=45)
plt.title('Number of Trips per Month', fontsize=15)
plt.show()
```

## Conclusion:

This analysis of Uber trip data provides key insights into trip purposes, duration, and locations. the strategic planning, and understanding user behavior.

###  Data preprocessing:
- The dataset was successfully cleaned by handling missing values in the 'PURPOSE' column and removing an anomalous 'Totals' row.
- Date columns ('START_DATE', 'END_DATE')are converted to datetime objects, and a 'Duration_Minutes' column was calculated, which proved crucial for time-based analysis.

###  Descriptive Statistics:
- **Miles:** The average trip distance is approximately **10.57 miles**, with a wide range (0.5 to 310.3 miles), indicating a mix of short and very long trips. The median of 6 miles suggests that most trips are relatively short.
- **Duration:** The average trip duration is about **19.33 minutes**. Similar to miles, there's a significant spread in duration (0 to 178 minutes), with a median of 15 minutes, again suggesting a skew towards shorter trips.

###  Trip Purposes:
- **Most Frequent Purposes:** 'Unknown' (due to initial missing data) was the most frequent, followed by 'Meeting', 'Meal/Entertain', 'Errand/Supplies', and 'Customer Visit'. This highlights a strong emphasis on business-related activities.
- **Miles and Duration by Purpose:** 'Commute' trips show the highest average miles (180.2 miles), indicating a few very long commuting trips. 'Customer Visit' (20.69 miles) and 'Meeting' (15.25 miles) also involve substantial distances. 'Customer Visit' and 'Meeting' trips also have longer average durations (around 31.6 minutes and 25.7 minutes, respectively).

###  Trip Categories:
- **Dominance of Business Trips:** The vast majority of trips (1078) are categorized as 'Business', with only 77 'Personal' trips. This suggests the dataset primarily captures professional usage of Uber.
- **Business vs. Personal Metrics:** Business trips have a slightly higher average distance (10.66 miles) and significantly longer average duration (19.61 minutes) compared to Personal trips (9.32 miles and 10.38 minutes). This might indicate that business trips involve more complex routes or multiple stops.

###  Locations:
- **Top Start and Stop Locations:** 'Cary', 'Unknown Location', 'Morrisville', and 'Whitebridge' consistently appear as the most frequent start and stop locations. 'Cary' in particular is a prominent hub for Uber activity in this dataset.


### Monthly Trends:
- The number of trips per month shows variation throughout the year, with potential peaks or troughs that could indicate seasonal usage patterns or specific business cycles. Further analysis could explore these trends in more detail.
